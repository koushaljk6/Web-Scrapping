```python
company =  soup.find_all('div', class_='mt-0')
len(company)
```

```python
name =[]
rating=[]
reviews=[]
salary=[]
jobsProvided=[]
empCount=[]
industry=[]


for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('b').text.strip())
    reviews.append(i.find('h3', attrs = {'data-test':'cell-Reviews-count'}).text.strip())
    salary.append(i.find('h3', attrs = {'data-test':'cell-Salaries-count'}).text.strip())
    jobsProvided.append(i.find('h3', attrs = {'data-test':'cell-Jobs-count'}).text.strip())
    empCount.append(i.find('span', attrs = {'data-test':'employer-size'}).text.strip())
    industry.append(i.find('span', attrs = {'data-test':'employer-industry'}).text.strip())
    
d={'Company_name':name, 'Company_rating':rating, 'Reviews':reviews, 'Salary':salary, 'Jobs_Provided':jobsProvided, 'Company_Size':empCount, 'Industry':industry}
df=pd.DataFrame(d)
df
```

```python
final=pd.DataFrame()
for j in range(1,31): #to collect data upto 30 pages
    url='https://www.glassdoor.co.in/Reviews/index.htm?overall_rating_low=3.5&page={}&filterType=RATING_OVERALL'.format(j)  #to make the link dynamic instead of page=1 we written page={} and also included .format()
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    company =  soup.find_all('div', class_='mt-0')
    name =[]
    rating=[]
    reviews=[]
    salary=[]
    jobsProvided=[]
    empCount=[]
    industry=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('b').text.strip())
        reviews.append(i.find('h3', attrs = {'data-test':'cell-Reviews-count'}).text.strip())
        salary.append(i.find('h3', attrs = {'data-test':'cell-Salaries-count'}).text.strip())
        jobsProvided.append(i.find('h3', attrs = {'data-test':'cell-Jobs-count'}).text.strip())
        empCount.append(i.find('span', attrs = {'data-test':'employer-size'}).text.strip())
        industry.append(i.find('span', attrs = {'data-test':'employer-industry'}).text.strip())

    d={'Company_name':name, 'Company_rating':rating, 'Reviews':reviews, 'Salary':salary, 'Jobs_Provided':jobsProvided, 'Company_Size':empCount, 'Industry':industry}
    df=pd.DataFrame(d)
    final=final.append(df, ignore_index=True)
```
